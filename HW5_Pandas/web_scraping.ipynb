```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

## Задание 1

```python
def habr_search(query):
    url = 'https://habr.com'
    search_url = f'{url}/ru/search/'
    posts_df = pd.DataFrame()
    
    for q in query:
        params = {'q': q}
        req = requests.get(search_url, params = params).text
        time.sleep(0.3)
        soup = BeautifulSoup(req)
        posts = soup.find_all('div', class_ = 'tm-article-snippet')
        for post in posts:
            date = post.find('span', class_ = 'tm-article-datetime-published').text
            title = post.find('h2', class_ = 'tm-title').text
            link = f"{url}{post.find('a', class_ = 'tm-title__link').get('href')}"
            row = {'search': q, 'date': date, 'title': title, 'link': link}
            posts_df = pd.concat([posts_df, pd.DataFrame([row])])
    
    posts_df.drop_duplicates(inplace = True)
    return posts_df.reset_index(drop = True)


################################
habr_search(['python', 'java', 'python'])
```

## Задание 2

```python
def habr_search_ext(query, pages = 1):
    url = 'https://habr.com'
    posts_df = pd.DataFrame()
    
    for page in range(1, pages + 1):
        search_url = f'{url}/ru/search/page{page}/'        
        
        for q in query:
            params = {'q': q}
            req = requests.get(search_url, params = params).text
            time.sleep(0.3)
            soup = BeautifulSoup(req)
            posts = soup.find_all('div', class_ = 'tm-article-snippet')
            
            for post in posts:
                date = post.find('span', class_ = 'tm-article-datetime-published').text
                title = post.find('h2', class_ = 'tm-title').text
                link = f"{url}{post.find('a', class_ = 'tm-title__link').get('href')}"
                req = requests.get(link).text
                time.sleep(0.3)
                soup = BeautifulSoup(req)
                text = soup.find('div', class_ = 'article-formatted-body')
                if text: 
                    text = text.text
                else: text = ''
                voices = soup.find('span', class_ = 'tm-votes-meter__value').text
                row = {'search': q, 'date': date, 'title': title, 'link': link, 'text': text, 'voices': voices}
                posts_df = pd.concat([posts_df, pd.DataFrame([row])])
    
    posts_df.drop_duplicates(inplace = True)
    return posts_df.reset_index(drop = True)


################################
habr_search_ext(['python'], 1)
```
